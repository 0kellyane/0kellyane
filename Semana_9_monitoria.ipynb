### Analisando as Tendências de Adesão de Alunos Internacionais

Você faz parte da equipe de análise de dados de uma universidade global que deseja entender as tendências de adesão de alunos internacionais ao longo dos últimos 20 anos. O objetivo é identificar os países com maior crescimento no número de matrículas e usar essas informações para direcionar as campanhas de marketing e adaptar o currículo para atrair mais estudantes internacionais. Para isso, vamos analisar os dados históricos de matrículas de vários países e compreender os padrões ao longo dos anos.

**Objetivo da Tarefa:**


O seu objetivo nesta atividade é analisar as tendências de adesão de alunos de diferentes países, focando especialmente nos alunos do Brasil. Você vai carregar e processar os dados, extrair as informações do Brasil e preparar tudo para visualizarmos as tendências de matrículas

```python
import pandas as pd
```

```python
alunos = pd.read_csv('/content/alunos_internacionais.csv')
alunos.head()
```

```python
#mudando os índices

alunos.set_index('País', inplace=True)

alunos
```

```python
anos = list(map(str, range(2000, 2021)))
```

```python
brasil = alunos.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'alunos': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head()
```

### Criando Gráficos para Analisar a Adesão de Alunos Internacionais

**Contextualização:**


Você faz parte da equipe de análise de dados de uma universidade global que deseja entender as tendências de adesão de alunos internacionais ao longo dos últimos 20 anos. Depois de analisar a tabela com os dados, a universidade deseja visualizar essas informações por meio de gráficos, para facilitar o entendimento das tendências ao longo do tempo.

**Objetivo da Tarefa:**


O seu objetivo nesta atividade é criar um gráfico de linha que mostre como a adesão de alunos brasileiros evoluiu de 2000 a 2020. Além disso, você irá ajustar as marcações dos eixos, melhorar a visualização, adicionar rótulos e um título ao gráfico.



```python
#importando o matplitlib

import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['alunos'])
plt.xticks(['2000', '2005', '2010', '2015', '2020'])
plt.xlabel('Ano')
plt.ylabel('Alunos')
```

#Criando Figuras e Personalizando Gráficos para Analisar a Adesão de Alunos Internacionais



Você faz parte da equipe de análise de dados de uma universidade global que deseja entender as tendências de adesão de alunos internacionais ao longo dos últimos 20 anos. Agora, vamos personalizar ainda mais o gráfico que você criou anteriormente, utilizando uma técnica mais avançada: a criação de figuras com a função subplots() da biblioteca Matplotlib. Isso vai permitir maior flexibilidade na hora de ajustar os eixos, adicionar rótulos e personalizar o gráfico.

**Objetivo da Tarefa:**


O seu objetivo é criar uma figura e personalizar o gráfico de adesão de alunos brasileiros, configurando os eixos e adicionando o título e os rótulos. Além disso, você vai aprender a controlar o intervalo das marcações no eixo x.


```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(dados_brasil['ano'], dados_brasil['alunos'])
ax.set_title(' Adesão de Alunos Internacionais')
ax.set_xlabel('Ano')
ax.set_ylabel('ALuno')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

# Comparando Dois Gráficos Lado a Lado – Gráfico de Linhas e Boxplot


Agora que você já entendeu como criar um gráfico de linha para mostrar as tendências de adesão de alunos brasileiros ao longo dos anos, vamos expandir a análise criando dois gráficos lado a lado. Isso nos permite comparar diferentes tipos de representações gráficas dos dados. Nesta atividade, você vai criar uma figura com dois gráficos: um gráfico de linha e um boxplot, que mostrará informações estatísticas sobre os dados.

Objetivo da Tarefa:


O seu objetivo é criar uma figura com dois gráficos lado a lado: um gráfico de linhas para mostrar a adesão dos alunos ao longo do tempo e um boxplot para visualizar a distribuição e amplitude desses dados. Você também vai adicionar títulos, rótulos, e grids (linhas de grade) para melhorar a clareza dos gráficos.

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(dados_brasil['ano'], dados_brasil['alunos'])
```

```python
#criando o dois gráficos definindo a quantidade de linhas no subplot e o tamannho da figura
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

#dados do segundo grafico
axs[0].plot(dados_brasil['ano'], dados_brasil['alunos'])
axs[0].set_title(' Adesão de Alunos Internacionais')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('ALuno')
# define a frequência como a cada 5 anos.
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
#adicionando grades ao gráfico
axs[0].grid()


#dados do segundo gráfico
axs[1].boxplot(dados_brasil['alunos'])
#titulo do gráfico
axs[1].set_title(' Adesão de Alunos Internacionais')
#rotulo de eixo x
axs[1].set_xlabel('Ano')
#rotulo do eixo y
axs[1].set_ylabel('Alunos')
#adicionando grades ao gráfico
axs[1].grid()


plt.show()
```

# Comparando Tendências de Adesão de Alunos de Países da América do Sul

Agora que você já sabe criar subplots para comparar dois gráficos lado a lado, vamos avançar e comparar as tendências de adesão de alunos dos quatro maiores países da América do Sul (Brasil, Colômbia, Argentina, e Peru) ao longo dos anos. O objetivo é entender se as tendências são parecidas entre esses países e se existem picos de adesão em períodos semelhantes ou diferentes.

Objetivo da Tarefa:


O seu objetivo é criar uma figura com quatro subplots (um para cada país) e organizar os gráficos em uma matriz de 2x2. Você também vai ajustar os eixos, adicionar rótulos e incluir um título geral para a figura.

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0,0].plot(alunos.loc['China', anos])
axs[0,0].set_title('China')

axs[0,1].plot(alunos.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(alunos.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(alunos.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

plt.show()
```

# Personalizando o Gráfico de Adesão de Alunos Brasileiros

Você faz parte da equipe de análise de dados de uma universidade e está responsável por gerar gráficos de adesão de alunos internacionais. Em um gráfico anterior, você já visualizou os dados de adesão de alunos brasileiros, mas o gráfico pode ser mais atrativo e fácil de interpretar. Agora, seu objetivo é personalizar o gráfico: ajustar o tamanho das fontes, melhorar a legibilidade com cores e espessura de linhas, adicionar grades, e entender como essas personalizações podem impactar a análise visual.

Objetivo da Tarefa:
O objetivo é modificar o gráfico de adesão de alunos brasileiros para melhorar sua visualização e legibilidade. Vamos alterar o tamanho das fontes, mudar a espessura da linha do gráfico, adicionar uma grade para facilitar a leitura dos valores e ajustar a posição do título. Ao final, você terá um gráfico mais atraente e informativo.

```python
fig, ax = plt.subplots(figsize=(19, 10))

ax.plot(dados_brasil['ano'], dados_brasil['alunos'], lw=1)
ax.set_title(' Adesão de Alunos Internacionais', fontsize= 18, fontweight = 'bold', family = 'serif', color='red', loc = 'left')
ax.set_xlabel('Ano', fontsize = 18)
ax.set_ylabel('Aluno', fontsize = 18 )
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_tick_params(labelsize=18)
ax.yaxis.set_tick_params(labelsize=18)
ax.grid(color='r', linestyle='-', linewidth=2)
```
